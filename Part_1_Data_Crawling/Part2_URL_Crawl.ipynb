```python
import requests
import json
import re
import pandas as pd
import time
```

```python
STATENAME = 'Pennsylvania'
names = pd.read_csv(STATENAME+'_documents.csv')['Name']
base_url = "https://customsearch.googleapis.com/customsearch/v1"
searched_name = []
links = []

for name in names:
    parameter_dictionary = {'cx':'', 'q': name+' '+STATENAME, 'key':''}
    response = requests.get(base_url, parameter_dictionary)
    result_dict = response.json()
    searched_name.append(result_dict['items'][0]['title'] if 'items' in result_dict else float("nan"))
    links.append(result_dict['items'][0]['link'] if 'items' in result_dict else float("nan"))
    print(result_dict['items'][0]['link'] if 'items' in result_dict else float("nan"))
    time.sleep(2)
```

```python
data={'Name':names, 'url':links}
pd.DataFrame(data).to_csv(STATENAME+'_urls.csv', index = False)
```
